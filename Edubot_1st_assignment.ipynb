```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
events = [
    ("1950", "Turing Test"),
    ("1956", "Dartmouth Conference"),
    ("1980s", "Expert Systems"),
    ("1997", "Deep Blue beats Kasparov"),
    ("2012", "AlexNet wins ImageNet"),
    ("2016", "AlphaGo beats Lee Sedol"),
    ("2018", "BERT released"),
    ("2023", "GPT-4 released")
]

years = [e[0] for e in events]
labels = [e[1] for e in events]

plt.figure(figsize=(12, 3))
plt.plot(years, [1]*len(years), "ro")
for i in range(len(labels)):
    plt.text(years[i], 1.00, labels[i], rotation=45, ha='right', fontsize=9)
plt.yticks([])
plt.title("Timeline of AI/ML Milestones")
plt.grid(True)
plt.show()
```

```python
# Create example data (X is input, y is output)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])  # y is just 2 times X

# Create the model
model = LinearRegression()

# Teach the model using our data
model.fit(X, y)

# Ask model to predict y values for X
predicted = model.predict(X)

# Show original data and predicted data on a graph
plt.scatter(X, y, color='blue', label='Actual values')
plt.plot(X, predicted, color='red', label='Predicted values')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression Example')
plt.legend()
plt.show()
```
